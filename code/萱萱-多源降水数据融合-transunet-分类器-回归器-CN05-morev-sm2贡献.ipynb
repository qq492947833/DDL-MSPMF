```python
import shap
import tensorflow as tf
import tensorflow.keras.backend as K
import numpy as np

# ==============================================================================
# 1. 定义 SHAP 辅助函数 (基于 Flatten + Wrapper 策略以节省显存)
# ==============================================================================

def create_simple_wrapped_model_for_shap(original_model, x_flat_len, output_channel_idx=None):
    """
    创建一个包装模型：
    输入: Flatten 后的向量 (Batch, H*W*C)
    操作: Reshape 回 (Batch, H, W, C) -> 原模型 -> 取指定通道 -> 求均值
    输出: 标量 (Batch, 1)
    """
    # 定义扁平化的输入层
    shap_input = tf.keras.Input(shape=(x_flat_len,), dtype=tf.float32)
    
    # 动态获取原模型的输入形状 (例如: 128, 160, 11)
    original_input_shape = original_model.input_shape[1:] 
    
    # 1. 将扁平输入 Reshape 回 4D 图片格式
    reshape_x = tf.keras.layers.Reshape(original_input_shape, name='reshape_x_simple')(shap_input)
    
    # 2. 传入原模型
    model_output = original_model(reshape_x)
    
    # 3. 处理输出，目标是将其转化为一个标量 (Scalar)
    # 如果输出是 4D (Batch, H, W, C_out)
    if K.ndim(model_output) == 4: 
        if output_channel_idx is not None:
            # 取出特定的输出通道 (例如降水通道)
            output_slice = model_output[..., output_channel_idx] 
            # Flatten 空间维度并求平均 -> (Batch,)
            scalar_output = tf.reduce_mean(tf.reshape(output_slice, (tf.shape(output_slice)[0], -1)), axis=1) 
        else:
            # 如果没有指定通道，对所有通道求平均 (通常用于单通道输出)
            scalar_output = tf.reduce_mean(tf.reshape(model_output, (tf.shape(model_output)[0], -1)), axis=1) 
    else: 
        # 如果模型输出已经是 2D 或其他形状
        scalar_output = tf.reduce_mean(tf.reshape(model_output, (tf.shape(model_output)[0], -1)), axis=1) 
    
    # 4. 扩展维度以符合 Keras 输出要求 (Batch, 1)
    final_output = tf.expand_dims(scalar_output, axis=-1)
    
    return tf.keras.Model(inputs=shap_input, outputs=final_output)

def compute_shap_values_for_single_model(current_model, current_testx, sample_x_shape, num_model_outputs):
    """
    对单个模型计算 SHAP 值
    """
    # 计算扁平化后的长度
    x_flat_len = int(np.prod(sample_x_shape))
    # 将输入数据扁平化，以喂给 GradientExplainer
    flat_testx = current_testx.reshape(len(current_testx), -1)
    
    processed_shap_x_outputs = []
    
    # 对每一个输出变量(通道)分别计算
    for output_ch_idx in range(num_model_outputs):
        print(f"正在计算第 {output_ch_idx+1}/{num_model_outputs} 个输出变量的 SHAP 值...")
        
        # 构建包装模型
        wrapped_model = create_simple_wrapped_model_for_shap(
            current_model, x_flat_len, output_channel_idx=output_ch_idx
        )
        
        # 使用 GradientExplainer
        # 注意：这里我们使用 flat_testx 作为背景集(background)和解释集(explain)
        # 显存优化：GradientExplainer 对 TF2 图模式支持较好
        explainer = shap.GradientExplainer(wrapped_model, flat_testx)
        
        # 计算 shap values
        shap_values_list = explainer.shap_values(flat_testx)
        
        # GradientExplainer 返回的是 list，取第一个 (对应唯一的输出)
        if isinstance(shap_values_list, list):
            raw_shap_values = shap_values_list[0]
        else:
            raw_shap_values = shap_values_list
        
        # 将计算出的扁平 SHAP 值 Reshape 回原始 4D 形状 (Samples, Lat, Lon, Channels)
        shap_x_reshaped = raw_shap_values.reshape(len(current_testx), *sample_x_shape)
        
        # 取绝对值 (因为我们关注贡献的大小，而不是方向)
        processed_shap_x_outputs.append(np.abs(shap_x_reshaped))
        
    return processed_shap_x_outputs

# ==============================================================================
# 2. 主执行逻辑
# ==============================================================================

# 配置参数 (模拟你的配置项)
ifweight = 'shap' # 开启 SHAP 计算
num_samples = 100 # 限制样本数量，防止显存爆炸

print("开始 SHAP 特征重要性分析...")
print(f"采样数量: {num_samples}")

# --- 1. 数据采样 ---
# 为了保护显存，我们只从 testx 中随机抽取 100 个样本进行解释
current_testx = testx
if testx.shape[0] > num_samples:
    # 固定随机种子以便复现
    np.random.seed(42)
    index = np.random.choice(testx.shape[0], num_samples, replace=False)
    current_testx = testx[index, ]
else:
    num_samples = testx.shape[0]

# --- 2. 准备参数 ---
sample_x_shape = current_testx.shape[1:] # (Lat, Lon, Channels)
num_model_outputs = testy.shape[3]       # 输出通道数 (通常为1)

# --- 3. 计算 SHAP ---
# 这里的 model 是你前面 load_model 加载好的模型
# 强制使用 CPU 进行 SHAP 计算通常更稳定 (避免 GPU OOM)，尽管慢一点
# 如果你想试 GPU，可以把 with tf.device 去掉，但在大模型上很容易炸
with tf.device('/CPU:0'):
    shap_outputs_per_channel = compute_shap_values_for_single_model(
        model, 
        current_testx, 
        sample_x_shape, 
        num_model_outputs
    )

# --- 4. 统计贡献度并打印 ---
# shap_outputs_per_channel 是一个 list，长度 = 输出通道数
# 每个元素 shape = (Samples, Lat, Lon, Input_Channels)

input_channel_count = current_testx.shape[-1]
weights = np.zeros((num_model_outputs, input_channel_count))

print('\n=== SHAP 贡献度计算结果 ===')

for i, output_shap_data in enumerate(shap_outputs_per_channel):
    # output_shap_data: (Samples, Lat, Lon, Input_Channels)
    
    # 对 (Samples, Lat, Lon) 维度求均值，得到每个输入特征的平均 SHAP 绝对值
    mean_contribution_per_feature = np.nanmean(output_shap_data, axis=(0, 1, 2))
    
    # 计算百分比
    total_contribution = np.nansum(mean_contribution_per_feature)
    
    if total_contribution != 0:
        percent_contribution = (mean_contribution_per_feature / total_contribution) * 100
    else:
        percent_contribution = np.zeros_like(mean_contribution_per_feature)
    
    weights[i, :] = percent_contribution
    
    # 打印结果
    for j in range(input_channel_count):
        print(f'预报因子 {j+1} 对预报值 {i+1} 的贡献：{percent_contribution[j]:.4f} ％')
    print('\n')

# 这里的 weights 矩阵和你之前 OOB 方法得到的格式一致，可以继续用于后续画图或保存
```

```python
def Auto_petal_plot(v,labels,selfcolor=None,labelsize=20,title=' ',ifsave='no',savename=None):
    import numpy as np
    import matplotlib.pyplot as plt

    plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams['axes.unicode_minus']=False

    if selfcolor is not None and isinstance(selfcolor, str):
        try:
            selfcolor = getattr(plt.cm, selfcolor)
        except AttributeError:
            selfcolor = plt.cm.viridis
    elif selfcolor is None:
         selfcolor = plt.cm.viridis

    v_array = np.array(v)
    labels_array = np.array(labels)

    if len(v_array) == 0:
        fig, ax = plt.subplots(subplot_kw={'polar': True}, figsize=(16,16), facecolor='white')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(title, y=1.05, fontsize=labelsize)
        plt.show()
        if ifsave == 'yes' and savename:
            fig.savefig(savename, dpi=300)
        return

    sorted_indices = np.argsort(v_array)[::-1]
    labels_sorted = labels_array[sorted_indices]
    v_values_sorted = v_array[sorted_indices]

    N = len(v_values_sorted)
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)

    min_val_orig = np.min(v_values_sorted)
    max_val_orig = np.max(v_values_sorted)
    range_val_orig = max_val_orig - min_val_orig

    scaled_min_display = 0.5
    scaled_max_display = 1.5
    if N == 0: 
        heights_for_plot = np.array([])
    elif range_val_orig < 1e-9:
        heights_for_plot = np.full_like(v_values_sorted, scaled_min_display, dtype=float)
    else:
        heights_for_plot = (v_values_sorted - min_val_orig) / range_val_orig * (scaled_max_display - scaled_min_display) + scaled_min_display

    gap = 0.1
    width = (2 * np.pi / N) * (1 - gap)

    fig, ax = plt.subplots(subplot_kw={'polar': True}, figsize=(16,16),facecolor = 'white')
    ax.spines['polar'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    if not(isinstance(selfcolor, list)):
        colors_to_use = selfcolor(np.linspace(0.2, 0.8, N))
    else:
        colors_to_use=selfcolor
    bars = ax.bar(theta, heights_for_plot, width=width, bottom=0.0, 
                  color=colors_to_use, 
                  edgecolor='none')

    max_plot_height_val = np.max(heights_for_plot) if N > 0 else 1.0
    label_radial_offset = max_plot_height_val * 0.2 +((labelsize-20)*0.005)

    for i in range(N):
        bar_plotted_height = heights_for_plot[i]
        original_value_for_label = v_values_sorted[i]
        current_label_text = labels_sorted[i]
        
        angle_rad_for_position = theta[i]

        r_label_position = bar_plotted_height + label_radial_offset
        
        angle_deg_of_petal = np.degrees(angle_rad_for_position)
        
        text_rotation_tangential = angle_deg_of_petal + 90.0
        final_text_rotation = text_rotation_tangential
        normalized_angle = final_text_rotation % 360
        if normalized_angle > 90 and normalized_angle < 270:
            final_text_rotation += 180.0
        
        ax.text(angle_rad_for_position, r_label_position, f'{current_label_text}: {original_value_for_label:.2f}',
                ha='center', va='center', fontsize=labelsize,
                rotation=final_text_rotation, rotation_mode='anchor')

    ax.set_theta_zero_location('N')
    ax.set_title(title,y = 1.2,fontsize = labelsize)
    plt.show()
    
    if ifsave == 'yes' and savename is not None:
        fig.savefig(savename, dpi=300)
```

```python
def Auto_petal_plot(v,labels,selfcolor=None,labelsize=20,title=' ',ifsave='no',savename=None):
    import numpy as np
    import matplotlib.pyplot as plt

    plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams['axes.unicode_minus']=False

    if selfcolor is not None and isinstance(selfcolor, str):
        try:
            selfcolor = getattr(plt.cm, selfcolor)
        except AttributeError:
            selfcolor = plt.cm.viridis
    elif selfcolor is None:
         selfcolor = plt.cm.viridis

    v_array = np.array(v)
    labels_array = np.array(labels)

    if len(v_array) == 0:
        fig, ax = plt.subplots(subplot_kw={'polar': True}, figsize=(16,16), facecolor='white')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(title, y=1.05, fontsize=labelsize)
        plt.show()
        if ifsave == 'yes' and savename:
            fig.savefig(savename, dpi=300)
        return

    sorted_indices = np.argsort(v_array)[::-1]
    labels_sorted = labels_array[sorted_indices]
    v_values_sorted = v_array[sorted_indices]

    N = len(v_values_sorted)
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)

    min_val_orig = np.min(v_values_sorted)
    max_val_orig = np.max(v_values_sorted)
    range_val_orig = max_val_orig - min_val_orig

    scaled_min_display = 0.5
    scaled_max_display = 1.5
    if N == 0: 
        heights_for_plot = np.array([])
    elif range_val_orig < 1e-9:
        heights_for_plot = np.full_like(v_values_sorted, scaled_min_display, dtype=float)
    else:
        heights_for_plot = (v_values_sorted - min_val_orig) / range_val_orig * (scaled_max_display - scaled_min_display) + scaled_min_display

    gap = 0.1
    width = (2 * np.pi / N) * (1 - gap)

    fig, ax = plt.subplots(subplot_kw={'polar': True}, figsize=(16,16),facecolor = 'white')
    ax.spines['polar'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    if not(isinstance(selfcolor, list)):
        colors_to_use = selfcolor(np.linspace(0.2, 0.8, N))
    else:
        colors_to_use=selfcolor
    bars = ax.bar(theta, heights_for_plot, width=width, bottom=0.0, 
                  color=colors_to_use, 
                  edgecolor='none')

    max_plot_height_val = np.max(heights_for_plot) if N > 0 else 1.0
    label_radial_offset = max_plot_height_val * 0.2 +((labelsize-20)*0.005)

    for i in range(N):
        bar_plotted_height = heights_for_plot[i]
        original_value_for_label = v_values_sorted[i]
        current_label_text = labels_sorted[i]
        
        angle_rad_for_position = theta[i]

        r_label_position = bar_plotted_height + label_radial_offset
        
        angle_deg_of_petal = np.degrees(angle_rad_for_position)
        
        text_rotation_tangential = angle_deg_of_petal + 90.0
        final_text_rotation = text_rotation_tangential
        normalized_angle = final_text_rotation % 360
        if normalized_angle > 90 and normalized_angle < 270:
            final_text_rotation += 180.0
        
        ax.text(angle_rad_for_position, r_label_position, f'{current_label_text}: {original_value_for_label:.2f}',
                ha='center', va='center', fontsize=labelsize,
                rotation=final_text_rotation, rotation_mode='anchor')

    ax.set_theta_zero_location('N')
    ax.set_title(title,y = 1.2,fontsize = labelsize)
    plt.show()
    
    if ifsave == 'yes' and savename is not None:
        fig.savefig(savename, dpi=300)
```

```python
def Auto_pie_plot(v, labels, selfcolor=None, labelsize=20, title=' ', ifsave='no', savename=None):

    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib.colors
    import math
    plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams['axes.unicode_minus']=False
    
    special_placement_rules = {}
    
    left_anchor_name = special_placement_rules.get('left_anchor')
    right_anchor_name = special_placement_rules.get('right_anchor')
    left_neighbor_names = special_placement_rules.get('left_neighbors', [])
    right_neighbor_names = special_placement_rules.get('right_neighbors', [])

    processed_color_input = selfcolor 
    if processed_color_input is not None and isinstance(processed_color_input, str):
        try:
            processed_color_input = getattr(plt.cm, processed_color_input)
        except AttributeError:
            processed_color_input = plt.cm.viridis
    elif processed_color_input is None:
         processed_color_input = plt.cm.viridis

    v_array = np.array(v)
    labels_array = np.array(labels)
    
    fig, ax = plt.subplots(figsize=(16,16), facecolor='white')

    if len(v_array) == 0:
        ax.axis('off') 
    else:
        valid_indices = v_array > 0
        v_positive = v_array[valid_indices]
        labels_positive = labels_array[valid_indices]

        if len(v_positive) == 0:
            ax.axis('off')
        else:
            initial_sorted_indices = np.argsort(v_positive)[::-1] 
            v_initial_sorted = v_positive[initial_sorted_indices]
            labels_initial_sorted = labels_positive[initial_sorted_indices]
            
            N = len(v_initial_sorted)
            
            colors_to_use = None
            if isinstance(processed_color_input, matplotlib.colors.Colormap):
                colors_to_use = processed_color_input(np.linspace(0.2, 0.8, N))
            elif isinstance(processed_color_input, list):
                colors_to_use = [processed_color_input[i % len(processed_color_input)] for i in range(N)] if len(processed_color_input) < N else processed_color_input[:N]
            else: 
                colors_to_use = plt.cm.viridis(np.linspace(0.2, 0.8, N))

            total_value_original = np.sum(v_initial_sorted)
            original_percentages = (v_initial_sorted / total_value_original) * 100.0 if total_value_original > 0 else np.zeros_like(v_initial_sorted)
            
            threshold_percentage = 5.0
            exponent = 0.6 
            v_for_pie_display = np.power(v_initial_sorted.astype(float) + 1e-9, exponent)

            pie_display_labels_text = [f'{label}: {value:.2f}' for label, value in zip(labels_initial_sorted, v_initial_sorted)]

            wedges, texts_pie_labels, texts_autopct_objects = ax.pie(
                v_for_pie_display,
                labels=pie_display_labels_text,
                colors=colors_to_use,
                autopct=lambda p: '', 
                startangle=90,      
                pctdistance=0.80,   
                labeldistance=1.05, 
                wedgeprops=dict(edgecolor='w')
            )

            plt.setp(texts_pie_labels, fontsize=int(labelsize * 0.7), wrap=True) 
            plt.setp(texts_autopct_objects, fontsize=int(labelsize * 0.6), color="white", weight="bold")

            small_slices_info_for_external = []
            
            anchor_label_objects = {}

            for i, wedge_obj in enumerate(wedges):
                if texts_autopct_objects and i < len(texts_autopct_objects):
                     texts_autopct_objects[i].set_text(f'{original_percentages[i]:.2f}%')
                
                current_label_name_on_pie = labels_initial_sorted[i]
                
                if texts_pie_labels[i].get_visible():
                    if current_label_name_on_pie == left_anchor_name:
                        anchor_label_objects['left'] = texts_pie_labels[i]
                    elif current_label_name_on_pie == right_anchor_name:
                        anchor_label_objects['right'] = texts_pie_labels[i]

                if original_percentages[i] < threshold_percentage and total_value_original > 0:
                    texts_pie_labels[i].set_visible(False)
                    if texts_autopct_objects and i < len(texts_autopct_objects):
                        texts_autopct_objects[i].set_visible(False)
                    
                    external_text = f'{labels_initial_sorted[i]}: {v_initial_sorted[i]:.2f}'
                    angle_deg_wedge_center = (wedge_obj.theta1 + wedge_obj.theta2) / 2.0
                    angle_deg_wedge_center = angle_deg_wedge_center % 360 

                    small_slices_info_for_external.append({
                        'wedge_obj': wedge_obj,
                        'text': external_text,
                        'original_value': v_initial_sorted[i], 
                        'angle_deg_center': angle_deg_wedge_center,
                        'label_name': labels_initial_sorted[i] 
                    })
            
            small_slices_info_for_external.sort(key=lambda x: x['original_value'])

            pie_radius = wedges[0].r if wedges else 1.0
            
            num_small_labels = len(small_slices_info_for_external)
            
            left_boundary_x = -pie_radius * 2.0 
            right_boundary_x = pie_radius * 2.0  
            fig.canvas.draw() 

            left_anchor_obj = anchor_label_objects.get('left')
            if left_anchor_obj:
                pos = left_anchor_obj.get_position()
                left_boundary_x = pos[0] + (pie_radius * 0.1 if pos[0] < 0 else -pie_radius * 0.3)

            right_anchor_obj = anchor_label_objects.get('right')
            if right_anchor_obj:
                pos = right_anchor_obj.get_position()
                right_boundary_x = pos[0] + (pie_radius * 0.3 if pos[0] < 0 else -pie_radius * 0.1)

            y_text_level = pie_radius * 1.20 
            title_y_position = 1.05 
            if num_small_labels > 0:
                title_y_position = 1.1 

            x_coords_final_for_small = np.zeros(num_small_labels)

            if num_small_labels > 0:
                max_x_deviation_of_text = pie_radius * 0.7 

                if num_small_labels == 1:
                     x_coords_final_for_small[0] = 0.0
                else:
                    x_coords_final_for_small = np.linspace(-max_x_deviation_of_text, 
                                                           max_x_deviation_of_text, 
                                                           num_small_labels)
                
                for k_idx, info_item in enumerate(small_slices_info_for_external):
                    current_x = x_coords_final_for_small[k_idx]
                    
                    if info_item['label_name'] in left_neighbor_names:
                        current_x = min(current_x, left_boundary_x - 0.03 * pie_radius)
                    elif info_item['label_name'] in right_neighbor_names:
                        current_x = max(current_x, right_boundary_x + 0.03 * pie_radius)
                    
                    x_coords_final_for_small[k_idx] = current_x
                
                left_neighbors_in_list = [
                    {'idx': k, 'x': x_coords_final_for_small[k]}
                    for k, info in enumerate(small_slices_info_for_external) if info['label_name'] in left_neighbor_names
                ]
                right_neighbors_in_list = [
                    {'idx': k, 'x': x_coords_final_for_small[k]}
                    for k, info in enumerate(small_slices_info_for_external) if info['label_name'] in right_neighbor_names
                ]

                if left_neighbors_in_list and right_neighbors_in_list:
                    rightmost_left_neighbor = max(left_neighbors_in_list, key=lambda item: item['x'])
                    leftmost_right_neighbor = min(right_neighbors_in_list, key=lambda item: item['x'])

                    if rightmost_left_neighbor['x'] > leftmost_right_neighbor['x']:
                        mid_point = (rightmost_left_neighbor['x'] + leftmost_right_neighbor['x']) / 2.0
                        gap = pie_radius * 0.05
                        x_coords_final_for_small[rightmost_left_neighbor['idx']] = mid_point - gap
                        x_coords_final_for_small[leftmost_right_neighbor['idx']] = mid_point + gap


            for i, info in enumerate(small_slices_info_for_external):
                current_text_x = x_coords_final_for_small[i]
                current_text_y = y_text_level
                
                wedge = info['wedge_obj']
                target_angle_on_wedge_deg = info['angle_deg_center']
                target_angle_rad = np.deg2rad(target_angle_on_wedge_deg)
                x_wedge_conn = pie_radius * np.cos(target_angle_rad)
                y_wedge_conn = pie_radius * np.sin(target_angle_rad)

                ax.annotate(info['text'],
                            xy=(x_wedge_conn, y_wedge_conn),
                            xytext=(current_text_x, current_text_y),
                            textcoords='data',
                            fontsize=int(labelsize * 0.6),
                            arrowprops=dict(arrowstyle="-",
                                            connectionstyle="arc3,rad=0",
                                            color='black',
                                            linestyle='--',
                                            shrinkA=5, 
                                            shrinkB=10),
                            bbox=dict(boxstyle="round,pad=0.25", fc="white", ec="gray", lw=0.5, alpha=0.9),
                            horizontalalignment='center', 
                            verticalalignment='center',
                            wrap=False)
            ax.axis('equal')  
    
    ax.set_title(title, y=title_y_position, fontsize=labelsize) 
    
    if ifsave == 'yes' and savename is not None:
        fig.savefig(savename, dpi=300, bbox_inches='tight')

    plt.show()
```
